### Methods
While there is an established correlation between trains and bus stops and displacements, greenways are often left without analysis as being a potential trigger for displacement. Boston contains several large greenways, and we attempt to analyze their impact by observing trends in income, homeownership, and poverty levels in the areas surrounding greenways.

Defining greenways as "off-street, landscaped, linear park spaces with a transportation component (like a multi-use path)," we identified three of Boston's largest and newest greenways:
- East Boston Greenway
- The Riverway
- Rose Fitzgerald Kennedy Greenway

Observing these trends requires data collected at a granular level, which is only available via the U.S. Census. The Census Bureau manages the American Community Survey, which contains community data and subsequent estimates down to a block level. The Census Bureau maintains the multi-year data to provide more statistically accurate insights on smaller communities and issues that might otherwise not have relevant or accurate data. The data is broken down on a tract level, so we used Boston's census tract breakdowns to identify the tracts in which the three greenways reside. These three greenways are within Suffolk and Norfolk county.

### Limitations

- Some of the tracts are missing data for certain years. For example, the data above is missing information for tracts 102.04 and 102.05 in 2018, yet has data for those tracts in later years. This may make it more difficult to analyze trends over time.
- The tracts might be too granular, and it may be difficult to seamlessly combine them to view the larger picture.
- Homeownership and income are metrics influenced by a multitude of other factors, so while it may be easy to develop a correlation between greenways and these statistics, it would be extremely difficult to establish causation.
- While some of this data is concrete, some of it comes from estimates from the Census Bureau. We have confidence in the Census Bureau to make accurate estimates, but nonetheless there is a margin of error that may skew our analysis.
- Many of Boston's greenspaces, like the Kennedy Greenway and Boston Commons, are very old and do not have data closer to their establishment. Furthermore, such data might not be relevant since the greenways are so old.
