```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
selected_columns = ['neighborhood', 'total_built_all_lanes', 'median_income', 'median_value']

# Filter the DataFrame to include only selected columns
selected_df = df[selected_columns]

# Convert 'neighborhood' to dummy variables
neighborhood_dummies = pd.get_dummies(selected_df['neighborhood'])

# Concatenate the dummy variables with the selected columns
selected_df = pd.concat([selected_df, neighborhood_dummies], axis=1)

# Drop the original 'neighborhood' column
selected_df = selected_df.drop('neighborhood', axis=1)

# Drop rows with missing values if needed
selected_df = selected_df.dropna()

# Create a correlation matrix
corr = selected_df.corr()

# Create a heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(corr, annot=True, cmap='RdBu', vmin=-1, vmax=1, fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix: Neighborhoods, Bike Lanes, and Socioeconomic Factors')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
selected_neighborhood = 'Dorchester'
selected_columns = ['total_built_all_lanes', 'median_income', 'median_value','median_monthly_rent','housed_population','owners_renters_ratio']

# Filter the DataFrame for Dorchester and select columns
dorchester_df = df[df['neighborhood'] == selected_neighborhood][selected_columns]

# Drop rows with missing values if needed
dorchester_df = dorchester_df.dropna()

# Create a correlation matrix
corr = dorchester_df.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='RdBu', vmin=-1, vmax=1, fmt=".2f", linewidths=.5)
plt.title(f'Correlation Matrix for {selected_neighborhood}: Bike Lanes and Socioeconomic Factors')
plt.show()
```

**Dorchester**

With the addition of bike lanes


1.   The Median income decreased
2.   The Property value decreased
3.   The Median rent decreased
2.   The Owners to Renters ratio decreased









```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
selected_neighborhood = 'Dorchester'
selected_columns = ['total_built_all_lanes', 'median_income', 'median_value', 'BFBL', 'BL', 'BLSL', 'SBL', 'SLM', 'SLMTC', 'SUP']

# Filter the DataFrame for Dorchester and selected years
dorchester_df = df[(df['neighborhood'] == selected_neighborhood) & (df['year'] >= 2010) & (df['year'] <= 2020)][selected_columns]

# Drop rows with missing values if needed
dorchester_df = dorchester_df.dropna()

# Create a correlation matrix
corr = dorchester_df.corr()

# Create a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, cmap='RdBu', vmin=-1, vmax=1, fmt=".2f", linewidths=.5)
plt.title(f'Correlation Matrix for {selected_neighborhood} (2010-2020): Bike Lanes, Socioeconomic Factors')
plt.show()
```

**Dorchester**

1.   The predominant types of lanes constructed were BL and SLM
